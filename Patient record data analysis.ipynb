 
Generate 1000 by 6 columns data using NumPy random method. Generate the patient_id , height, weight, sex and age columns separately. Calculate their bmi = weight/height**2 . Using NumPy, stack all columns together. Transform your data into pandas DataFrame. Create a new column in your Dataframe and name it “interpretation”. Fill the column with underweight, normalweight, overweight, obese, clinicallyobese based on their bmi column. Using your dataset perform statistical analysis and answer the following




•Mean height, weight and age

•All underweight and obese

•How many underweight patient are 30 years but below 45

•How many patient are normal and are females below 25 years

•Number of males who are obese 

•number of male and females who are above the average age and are normal.


```python
# steps:

# Import Numpy package
# Random number generation
# Stack all the columns together
# Transform into panda dataframes
# Add another column and assign values
# Calculations
```

```python
#Import NumPy Package
import numpy as np
rng = np.random.default_rng(seed =1) #seed ensures reproducibility
```

#### Random Number Generation

```python
#Generating array for patient_id, age , sex, weight,height

# 1.Patient_id is chosen as a 3-digit integer
patient_id = np.arange(100,1100)


# 2.Age is chosen between 1 and 100 inclusive
Age = rng.integers(low=1, high=100, size=1000)

# 3.Sex is chosen as an integer (1= Male, 0=Female)
Sex = rng.integers(low=0, high=1, size=1000, endpoint =True)


#4. Weight is a float value 30-200
weight = np.around(rng.uniform(low=30, high=200, size=1000),2)

#5. Height is a float value 1-1.9
height = np.around(rng.uniform(low=1, high=1.9, size=1000),2)
# BMI calculations
BMI = weight/height**2
```

#### Stacking all the columns together

```python
#use np.column_stack()
patient_record = np.column_stack([patient_id,Age, Sex, weight,height, BMI])
print(patient_record)
```

#### Transform the data into panda dataframes

```python
#Import pandas
import pandas as pd
#Create Dataframe

pat_rec = pd.DataFrame(patient_record)
pat_rec
#Rename the columns
pat_rec.columns=['patient_id', 'Age','Sex', 'weight','height', 'BMI']
pat_rec

#Reassign the datatypes for the columns
pat_rec=pat_rec.astype({'patient_id': int, 'Age': int,  'Sex':int})

pat_rec
```

#### Adding another column to the dataframe
#### Assign Interpretation

```python
#1. Create an empty list Interpretation = []

Interpretation = []

#2. Use a for loop to assign the values based on data in BMI
#Key: less than 18 => Underweight; between 18.5 and 24.99 inclusive=> NormalWeight;
    # between 25 and 29.99 inclusive=> OverWeight; between 30 and 34.99 inclusive=> Obese
    # 35 and above => clinically Obese

for i in  pat_rec['BMI']:
    if i<18.5:
        Interpretation.append('Underweight')
    elif i>=18.5 and i<25:
        Interpretation.append('Normal Weight')
    elif i>=25 and i <30:
        Interpretation.append('OverWeight')
    elif i>=30 and i<35:
        Interpretation.append('Obese')
    elif i>35:
        Interpretation.append('Clinically Obese')
                        


#3. Create new column by assignment    

pat_rec['Interpretation']=Interpretation
pat_rec
```
